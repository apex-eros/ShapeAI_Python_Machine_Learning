**WE GENERALLY START THE CODE BY IMPORTING LIBRARIES WHICH WE WILL USE THROUGHOUT THE PROGRAM.

```python
import numpy as np
import pandas as pd
import sklearn
```

**Now we will load the data that we will be using to train our model**

Here we are using boston house pricing dataset which is available in the sklearn library itself.


```python
from sklearn.datasets import load_boston
df = load_boston()
```

```python
df.keys()  #Return all the keys of the dataset dictionary
```

```python
print(df.data)  #Info about the dataset
```

**We convert our dataset into the pandas dataframe, so that it is easier to analyse the data**

```python
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head()
```

**Adding a new column of target values to the dataframe **

```python
boston['PREDICT'] = df.target
boston.head()
```

**Check if the dataset contains any null value or not**

```python
boston.isnull()
```

```python
boston.isnull().sum()
```

**We never train the model on all the data that we have, we always make sure to have a test dataset, which is different from the training dataset.**

```python
from sklearn.model_selection import train_test_split
X = boston.drop('PREDICT',axis = 1)
Y = boston['PREDICT']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size = 0.15, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

**Now lets import the Linear Regression Model from sklearn and train it on the training dataset**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# FITTING MODEL ON THE TRAINING DATASET

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize = True)
```

```python
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train,y_train_predict)))
print("The model performance for training dataset")
print('RMSE is {}'.format(rmse))
print("\n")

# ON TESTING DATASET

y_train_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test,y_train_predict)))
print("The model performance for testing dataset")
print('RMSE is {}'.format(rmse))


```
